```python
import numpy as np
import matplotlib.pyplot as plt

hbar = 1.0
I = 1.0
ells = np.arange(-8, 9)
phis = [0.0, 0.5*np.pi, np.pi, 1.5*np.pi]

plt.figure(figsize=(7, 4.5))
for phi in phis:
    E = (hbar**2/(2*I)) * (ells - (phi/(2*np.pi)))**2
    plt.plot(ells, E, marker='o', label=f'={phi:.2f}')
plt.xlabel('')
plt.ylabel('E_')
plt.title('Rotor spectrum shift with holonomy')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
